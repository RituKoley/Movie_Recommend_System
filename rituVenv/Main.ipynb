```python
# In Python, the ast module (short for Abstract Syntax Trees) allows you to interact with the syntactic structure of Python code programmatically. Here's a breakdown:
# What is an AST?
# An Abstract Syntax Tree is a tree-like representation of the structure of a program. Each node in the tree represents a construct occurring in the code, such as a variable, function, or control flow statement. 
# Why is AST useful?
# Code analysis:
# You can use ASTs to analyze code structure, extract information, and identify patterns.
# Code manipulation:
# You can modify the AST and then generate new Python code from it. This allows you to:
# Transform code automatically
# Implement custom code transformations
import ast
```

```python
# converting by name
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
# genres - [{"id": 28, "name": "Action"}, {"id": 12, "name...
# keywords-	[{"id": 1463, "name": "culture clash"}, {"id":...
# convert them into name types like see the output
movies_df['genres']=movies_df['genres'].apply(convert)
movies_df['keywords']=movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
# convert cast and crew into name types like see the output
# converting by name
def convert3(obj):
    l = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
            counter+=1
        else:
            break
        return l
```

```python
movies_df['cast']=movies_df['cast'].apply(convert3)
movies_df.head()
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies_df['crew']=movies_df['crew'].apply(fetch_director)
movies_df.head()
```

```python
# feature extraction using count vectorizer
# sklearn.feature_extraction.text is a module in scikit-learn that provides tools to extract features from text data. 
# CountVectorizer: A text preprocessing tool that converts a collection of text documents to a matrix of token counts (bag-of-words representation).
# max_features=5000: Limits the vocabulary to the 5,000 most frequent words across the documents.
# stop_words='english': Removes common English words (e.g., "the", "and", "is") that are unlikely to add much semantic value to the analysis.
# This helps focus on more meaningful words.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
# convert vectors to array
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# nltk = natural language toolkit , a platform use to build pyhton program, 
# The Natural Language Toolkit (NLTK) is a Python programming environment for creating applications for statistical natural language processing (NLP). It includes language processing libraries for tokenization, parsing, classification, stemming, labeling, and semantic reasoning.
import nltk
```

```python
# PorterStemmer, a tool in the NLTK library used for stemming words (reducing words to their root forms).
# 
from nltk.stem.porter import PorterStemmer
# Create an Instance:
ps = PorterStemmer()
```

```python
# apply stemming in tags
def stem(text):
    y=[]
    for i in text.split():
        # Apply Stemming: stem words using the stem() method
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# The cosine_similarity function from sklearn.metrics.pairwise is used to calculate the cosine similarity between two vectors or a set of vectors.
#  Cosine similarity is a measure that calculates the cosine of the angle between two non-zero vectors in a multi-dimensional space, often used in text analysis, recommendation systems, and clustering.
# here we sue check the text similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
# similar col and rows
cosine_similarity(vectors).shape
```

```python
# assign cosine_similarity() into variable similarity  
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
# create sorted list
# enumerate() = convert data collection object to enumerate object, returns an object as counter of a key for each value in a object
sorted(
    list(enumerate(similarity[0])),  # Pair each similarity score with its index
    reverse=True,                   # Sort in descending order (highest similarity first)
    key=lambda x: x[1]              # Sort by the similarity score (second element in the tuple)
)[1:6]                              # Exclude the top result and take the next 5 results

# Step-by-Step Breakdown
# enumerate(similarity[0]):

# similarity[0] is assumed to be a list of similarity scores for a specific item (e.g., a document, vector, or word).
# enumerate() adds indices to each similarity score, returning something like:
# python
# Copy code
# [(0, 1.0), (1, 0.85), (2, 0.72), ...]
# list():

# Converts the enumerated object into a list of tuples.
# sorted():

# Sorts the list of tuples. Each tuple has the format (index, similarity_score).
# key=lambda x: x[1]: Sorts based on the similarity score (x[1]).
# reverse=True: Sorts in descending order (highest scores first).
# [1:6]:

# Excludes the first item (index 0), which is typically the self-similarity of the item (always 1.0 in cosine similarity).
# Selects the next top 5 results.
```

```python
# create recommend function
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),
                         reverse=True,
                         key= lambda x:x[1]
                         )[1:6]
    # for i in movies_list:
    #     print(new_df.iloc[i[0]]['title'])
    recommended_movies = [new_df.iloc[i[0]]['title'] for i in movies_list]
    return recommended_movies
```
